```python
# Digitized categorical variables
encoders = {}
to_encode_columns = []
for column in df.columns:
    if df[column].dtype == "object" or df[column].dtype == "bool":
        to_encode_columns.append(column)
for column in to_encode_columns:
    encoder = LabelEncoder()
    col = df[column]
    encoded_col = encoder.fit_transform(col)
    df[column] = encoded_col
    encoders[column] = encoder
```

```python
# set the part 
X = df[["transmission", "odometer_value", "year_produced", "drivetrain", "number_of_photos"]]
y = df[["price_usd"]]
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 2022)
```

```python
# build the model and predict the price
rf = RandomForestRegressor()
rf.fit(X_train, Y_train)
```

```python
Y_hat = rf.predict(X_test)
r2 = metrics.r2_score(Y_test, Y_hat)
r2
```

```python
important_features = pd.Series(data = rf.feature_importances_,index = X_train.columns)
important_features.sort_values(ascending = False, inplace = True)
important_features
```

```python
# Users can interact with the program, enter data and get predicted prices
def predict_price(model = rf, data = X):
    feature = pd.DataFrame(columns = data.columns)
    transmission = input("please input the transmission(mechanical or automatic)")
    transmission = encoders["transmission"].transform([transmission])[0]
    feature["transmission"] = [transmission]
    odometer_value = input("please input the odometer value")
    feature["odometer_value"] = [odometer_value]
    year_produced = input("please input the year produced")
    feature["year_produced"] = [year_produced]
    drivetrain = input("please input the drivetrain")
    drivetrain = encoders["drivetrain"].transform([drivetrain])[0]
    feature['drivetrain'] = [drivetrain]
    number_of_photos = input("please input the number of photos")
    feature["number_of_photos"] = [number_of_photos]
    price = model.predict(feature)
    return price[0]
```

```python
predict_price()
```

```python
# Build other models to predict prices
ab = AdaBoostRegressor()
ab.fit(X_train, Y_train)
Y_hat = ab.predict(X_test)
r2 = metrics.r2_score(Y_test, Y_hat)
r2
```

```python
# Build other models to predict prices
lin = LinearRegression()
lin.fit(X_train, Y_train)
Y_hat = lin.predict(X_test)
r3 = metrics.r2_score(Y_test, Y_hat)
r3
```

```python
# Build other models to predict prices
gb = GradientBoostingRegressor()
gb.fit(X_train, Y_train)
Y_hat = gb.predict(X_test)
r4 = metrics.r2_score(Y_test, Y_hat)
r4
```
